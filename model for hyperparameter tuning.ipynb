```python
import optuna
from catboost import CatBoostClassifier
from sklearn.metrics import f1_score
from sklearn.model_selection import cross_val_score, StratifiedKFold
import numpy as np

# Feature와 Target 분리
X = dispenser1_df_train.drop(columns='target')
y = dispenser1_df_train['target']

# Optuna Objective Function 정의
def objective(trial):
    param = {
        'learning_rate': trial.suggest_loguniform('learning_rate', 1e-2, 0.1),
        'depth': trial.suggest_int('depth', 7, 10),
        'l2_leaf_reg': trial.suggest_loguniform('l2_leaf_reg', 1e-3, 0.1),
        'n_estimators': trial.suggest_int('n_estimators', 300, 500),
        'task_type': 'GPU'
    }

    model = CatBoostClassifier(
        **param,
        loss_function='Logloss',
        eval_metric='F1',
        cat_features=["Chamber Temp. Judge Value_AutoClave", "Change in the Column"],
        verbose=0
    )

    # Stratified K-Fold Cross Validation
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state= 110)
    f1_scores = []

    for train_index, valid_index in skf.split(X, y):
        X_train, X_valid = X.iloc[train_index], X.iloc[valid_index]
        y_train, y_valid = y.iloc[train_index], y.iloc[valid_index]
        
        model.fit(X_train, y_train, eval_set=(X_valid, y_valid), early_stopping_rounds=200, verbose=False)
        
        # Validation prediction and F1 score calculation
        y_valid_pred = model.predict(X_valid)
        f1 = f1_score(y_valid, y_valid_pred)
        f1_scores.append(f1)
    
    # Return the mean F1 score across all folds
    return np.mean(f1_scores)

# Optuna Study 생성 및 최적화
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials= 200)

# 최적의 하이퍼파라미터 출력
print("Best Parameters:", study.best_params)
print("Best F1 Score:", study.best_value)
```

```python
# 최적의 하이퍼파라미터로 모델 학습
best_params = study.best_params
best_model = CatBoostClassifier(
    **best_params,
    loss_function='Logloss',
    eval_metric='F1',
    cat_features=["Chamber Temp. Judge Value_AutoClave", "Change in the Column"],
    verbose=0
)

# 전체 훈련 데이터에서 모델을 재학습
best_model.fit(X, y, early_stopping_rounds=200, verbose=False)

# 교차 검증으로 최종 F1 Score 계산
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
final_f1_scores = []

for train_index, valid_index in skf.split(X, y):
    X_train, X_valid = X.iloc[train_index], X.iloc[valid_index]
    y_train, y_valid = y.iloc[train_index], y.iloc[valid_index]
    
    best_model.fit(X_train, y_train, eval_set=(X_valid, y_valid), early_stopping_rounds=150, verbose=False)
    
    # Validation prediction and F1 score calculation
    y_valid_pred = best_model.predict(X_valid)
    f1 = f1_score(y_valid, y_valid_pred)
    final_f1_scores.append(f1)

# 최종 F1 Score 출력
final_f1_score = np.mean(final_f1_scores)
print("Best Parameters:", best_params)
print("Final F1 Score:", final_f1_score)
```

```python
import optuna
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import StratifiedKFold
import numpy as np

# Feature와 Target 분리
X = dispenser2_df_train.drop(columns='target')
y = dispenser2_df_train['target']

# Optuna Objective Function 정의
def objective(trial):
    param = {
        'learning_rate': trial.suggest_loguniform('learning_rate', 1e-2, 0.1),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 10),
        'gamma': trial.suggest_loguniform('gamma', 1e-3, 0.1),
        'subsample': trial.suggest_uniform('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_uniform('colsample_bytree', 0.5, 1.0),
        'alpha': trial.suggest_loguniform('alpha', 1e-3, 0.1),
        'lambda': trial.suggest_loguniform('lambda', 1e-3, 0.1),
        'scale_pos_weight': trial.suggest_loguniform('scale_pos_weight', 1, 10),
        'n_estimators': trial.suggest_int('n_estimators', 300, 500),
        'objective': 'binary:logistic',
        'eval_metric': 'auc',  # AUC로 변경
        'tree_method': 'gpu_hist',  # GPU 지원
    }

    model = XGBClassifier(
        **param,
        use_label_encoder=False,
        verbose=0
    )

    # Stratified K-Fold Cross Validation
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=110)
    auc_scores = []

    for train_index, valid_index in skf.split(X, y):
        X_train, X_valid = X.iloc[train_index], X.iloc[valid_index]
        y_train, y_valid = y.iloc[train_index], y.iloc[valid_index]
        
        model.fit(
            X_train, y_train, 
            eval_set=[(X_valid, y_valid)],
            verbose=False
        )
        
        # Validation prediction and AUC score calculation
        y_valid_pred_prob = model.predict_proba(X_valid)[:, 1]
        auc = roc_auc_score(y_valid, y_valid_pred_prob)
        auc_scores.append(auc)
    
    # Return the mean AUC score across all folds
    return np.mean(auc_scores)

# Optuna Study 생성 및 최적화
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=300)

# 최적의 하이퍼파라미터 출력
print("Best Parameters:", study.best_params)
print("Best AUC Score:", study.best_value)
```

```python
# 최적의 하이퍼파라미터로 모델 학습
best_params = study.best_params
best_model = XGBClassifier(
    **best_params,
    use_label_encoder=False,
    verbose=0
)

# 전체 훈련 데이터에서 모델을 재학습
best_model.fit(X, y, eval_set=[(X, y)], verbose=False)

# 교차 검증으로 최종 F1 Score 계산
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
final_f1_scores = []

for train_index, valid_index in skf.split(X, y):
    X_train, X_valid = X.iloc[train_index], X.iloc[valid_index]
    y_train, y_valid = y.iloc[train_index], y.iloc[valid_index]
    
    best_model.fit(X_train, y_train, eval_set=[(X_valid, y_valid)], verbose=False)
    
    # Validation prediction and F1 score calculation
    y_valid_pred = best_model.predict(X_valid)
    f1 = f1_score(y_valid, y_valid_pred)
    final_f1_scores.append(f1)

# 최종 F1 Score 출력
final_f1_score = np.mean(final_f1_scores)
print("Best Parameters:", best_params)
print("Final F1 Score:", final_f1_score)
```

```python
# 예측 확률에 AbNormal로 뺴두었던 index 추가하기

def add_unconditionally_AbNormal_dispenser1(pred_proba_numpy_array, val_x, submission_abnormal_row_index):
  # y_pred_proba1을 DataFrame으로 변환
  y_pred_proba_df = pd.DataFrame(y_pred_proba1, index=val_x_dispenser1.index, columns=['probability'])

  # Submission_AbNormal_row_index의 인덱스를 y_pred_proba_df에 추가하고, 값을 1로 설정
  for idx in submission_abnormal_row_index:
    if idx not in y_pred_proba_df.index:
      y_pred_proba_df.loc[idx] = 1.0  # 인덱스가 없으면 추가하고 값은 1.0으로 설정

  # y_pred_proba_df을 다시 Numpy로 변경
  modified_pred_proba_array = y_pred_proba_df.sort_index()['probability'].values

  return modified_pred_proba_array
```

```python
# validatoin_y에 AbNormal 추가하기
def add_abnormal_rows(val_y, submission_abnormal_row_index):
  # val_y의 기존 인덱스
  existing_indices = val_y.index.tolist()
  # Submission_AbNormal_row_index에서 val_y의 인덱스와 겹치지 않는 인덱스를 필터링
  new_indices = [idx for idx in submission_abnormal_row_index if idx not in existing_indices]

  # 기존 데이터프레임에 새로운 인덱스를 추가하고 'target' 값을 1로 설정
  for idx in new_indices:
    val_y.loc[idx] = 1
  # 인덱스를 정렬
  updated_val_y = val_y.sort_index()

  # 인덱스를 정렬
  updated_val_y = updated_val_y.values

  return updated_val_y
```
